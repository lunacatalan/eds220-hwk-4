```python
# import libraries
import os 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import xarray as xr
import rioxarray as rioxr
import geopandas as gpd

from shapely.geometry import box
from shapely.geometry import Point

import matplotlib.patches as mpatches # to create legends
import matplotlib.lines as mlines # to create legend shapes

## ===== Import Data =====

# import land cover data
lulc_fp = os.path.join(os.getcwd(), 
                      'data',
                      'land_cover.tif')

# open file
lulc = rioxr.open_rasterio(lulc_fp)

# import california shapefile with url
ca = gpd.read_file('https://data.ca.gov/dataset/e212e397-1277-4df3-8c22-40721b095f33/resource/3db1e426-fb51-44f5-82d5-a54d7c6e188b/download/ca-state-boundary.zip')

## ===== Make Bounding Box =====

# make bounds into bbox
bbox = gpd.GeoDataFrame({"id":1,
                         "geometry":[box(*lulc.rio.bounds())]}, # create lulc bbox by calling bounds 
                       crs = "epsg:5070") # same crs as lulc


## ===== Make Mt.Whitney Point =====

# mt whitney coordinates 36.5785Â° N, 118.2923Â° W
mt_whitney = Point( -118.2923, 36.5785)

mt_whitney = gpd.GeoDataFrame(geometry = [mt_whitney],
                             crs = "epsg:4326") # the crs that google map is using

## ===== Make Updates =====

# remove band dimension from lulc
lulc = lulc.squeeze().drop('band')

# update crs of ca and mt. whitney
ca = ca.to_crs("epsg:5070")
mt_whitney = mt_whitney.to_crs("epsg:5070")

## ===== Plot =====

# figure set up 
fig, ax = plt.subplots(figsize=(10,8))
ax.axis('off')

# to plot the ca shape
ca.plot(ax = ax,
       color = "khaki", # fill in 
       edgecolor = "black") # border around shapefile
# to add ca to legend
ca_patch = mpatches.Patch(color = "khaki",
                          label = "California, CA")

# add the box defined above to the map
bbox.plot(ax = ax,
          edgecolor = "black",
         color = "green") 
# add box to the legend
bbox_patch = mpatches.Patch(color = "green",
                           label = "LULC tile")

# add the point defined above 
mt_whitney.plot(ax = ax,
               color = "red",
               marker = 6) # make marker a triangle
# add the point to the legend
mt_whitney_patch = mlines.Line2D([],[],
                                 color = "red",
                                 marker = 6,
                                 label = "Mount Whitney",
                                 lw = 0,
                                 markersize = 9)

# edit legend
ax.legend(handles = [ca_patch, bbox_patch, mt_whitney_patch], # which labels to show 
          frameon = False, # no border around legend
          loc=(0.6, 0.8)) # where legend is located

plt.show()
```
